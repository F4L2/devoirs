```python
import pandas as pd
import numpy as np
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import pyAgrum.lib.bn2graph as bnPlot
import pyAgrum.lib.dynamicBN as gdyn
```

## Simulation/Tutorial

Empty net

```python
bn=gum.BayesNet('Diabete Remission')
print(bn)
```

Add nodes

```python
gly =bn.add(gum.LabelizedVariable('gly','glycemia > 6.5',2))
ins =bn.add(gum.LabelizedVariable('ins','insulin intake',2))
dru =bn.add(gum.LabelizedVariable('dru','other drugs intake',2))
rem =bn.add(gum.LabelizedVariable('rem','remission',2))
gly, ins, dru, rem
```

```python
bn
```

Connect nodes

```python
bn.addArc( gly, ins)
bn.addArc( gly, dru)
bn.addArc( dru, rem)
bn.addArc( ins, rem)
```

```python
bn
```

```python
bn.cpt(gly).fillWith([0.5,0.5])
```

```python
# order is [False, True] 
bn.cpt(ins)[:]=[ [0.1, 0.9],
                [0.9, 0.1] ]

bn.cpt(ins)[0], bn.cpt(ins)[1]
```

```python
# order is [False, True] 
bn.cpt(dru)[:]=[ [0.7, 0.25],
                [0.3, 0.75] ]

bn.cpt(dru)[0], bn.cpt(dru)[1]
```

```python
bn.cpt(rem)[0,0,:] = [0.1, 0.9]
bn.cpt(rem)[0,1,:] = [0.4, 0.6]
bn.cpt(rem)[1,0,:] = [0.7, 0.3]
bn.cpt(rem)[1,1,:] = [0.9, 0.1] 
```

```python
ie=gum.LazyPropagation(bn)
```

```python
ie.makeInference()
ie.posterior(rem)
```

```python
gnb.showInference(bn,evs={})
```

#### Glycemia < 6.5 && No drugs taken

```python
gnb.showInference(bn,evs={'gly':0,'dru':0})
```

36% remission in these conditions

#### Glycemia > 6.5 && insulin intake

```python
gnb.showInference(bn,evs={'gly':1,'ins':1})
```

15.71% of remission in these condition

## Bayesian network on real Data

### Discretize data

```python
df_env = pd.read_csv("data/SPLEX_env.txt", sep=' ')
df_host = pd.read_csv("data/SPLEX_host.txt", sep=' ')

df_host.head()
```

```python
l=[]
for col in df_env.columns.values:
    bins = np.linspace(min(df_env[col]), max(df_env[col]), 5)
    l.append(pd.DataFrame(np.digitize(df_env[col], bins),columns=[col]))

discr_data=pd.concat(l, join='outer', axis=1)
discr_data.to_csv("data/discr_env.csv",index=False)


l=[]
for col in df_host.columns.values:
    bins = np.linspace(min(df_host[col]), max(df_host[col]), 5)
    l.append(pd.DataFrame(np.digitize(df_host[col], bins),columns=[col]))

discr_data=pd.concat(l, join='outer', axis=1)
discr_data.to_csv("data/discr_host.csv",index=False)
```

### Environment dataset

```python
import os
if not os.path.exists('plots'):
    os.mkdir("plots")
    
df_env = pd.read_csv("data/discr_env.csv")
#df_host = pd.read_csv("data/discr_host.csv")

df_env.head()
```

```python
learner=gum.BNLearner("data/discr_env.csv")
learner.useLocalSearchWithTabuList()

bn=learner.learnBN()
bnPlot.dotize(bn, 'plots/Tabu_env', format='pdf')
```

```python
learner=gum.BNLearner("data/discr_env.csv")
learner.useGreedyHillClimbing()

bn=learner.learnBN()
bnPlot.dotize(bn, 'plots/HC_env', format='pdf')
```

### Host dataset

```python
learner=gum.BNLearner("data/discr_host.csv")
learner.useLocalSearchWithTabuList()

bn=learner.learnBN()
bnPlot.dotize(bn, 'plots/Tabu_host', format='pdf')
```

```python
learner=gum.BNLearner("data/discr_host.csv")
learner.useGreedyHillClimbing()

bn=learner.learnBN()
bnPlot.dotize(bn, 'plots/HC_host', format='pdf')
```

### Both dataset

```python
#We know they have the same number of observation here, so we just concat.

both = pd.concat([df_env, df_host], axis=1)
both.to_csv("data/discr_both.csv",index=False)
both.head()
```

```python
learner=gum.BNLearner("data/discr_both.csv")
learner.useLocalSearchWithTabuList()

bn=learner.learnBN()
bnPlot.dotize(bn, 'plots/Tabu_both', format='pdf')
```

```python
learner=gum.BNLearner("data/discr_both.csv")
learner.useGreedyHillClimbing()

bn=learner.learnBN()
bnPlot.dotize(bn, 'plots/HC_both', format='pdf')
```

Both networks are very similar if we infer from separate datasets, the connections are not the same but the connected nodes are close.   
But when we concatenate both dataset and infer the network from it, the networks become way less similar, the connected nodes are different.
