## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn import svm

from scipy.io import arff
```

### Cross-validation functions

```python
def split_data(data, k):
    chunks = [None]*k
    c_size = int(len(data) / k)
    for i in range(k):
        chunks[i] = data[i*c_size:(i+1)*c_size]
    return chunks

def cross_validation(X,Y, k, model):
    data = list(zip(X,Y))
    chunks = split_data(data,k)
    scores = []
    for i in range(len(chunks)):
        test_data = chunks[i].copy()
        test_X, test_y = zip(*test_data)

        train_chunks = [chunks[c].copy() for c in range(len(chunks)) if(c != i)]
        train_data = []
        for tc in train_chunks:
            train_data.extend(tc)
            
        train_X, train_y = zip(*train_data)

        model.fit(train_X, train_y)
        prediction = model.predict(test_X)

        score = sum( [1 if(y == y_est) else 0 for y,y_est in zip(test_y, prediction)] ) / len(test_y)
        scores.append(score)
    
    return np.mean(scores), scores
```

## Simulation data

### Generate simulated datasets

```python
#generate simulated datasets
X1, Y1 = make_classification(n_samples=10000, n_features=2, n_redundant=0, n_informative=2,n_clusters_per_class=1)
X2, Y2 = make_blobs(n_samples=10000, n_features=2, centers=2) #should be 3 centers but since SVM is binary, we'll reduce to 2 for simplicity's sake
X3, Y3 = make_moons(n_samples=10000, shuffle=True, noise=None, random_state=None)

Xs = [X1,X2,X3]
Ys = [Y1,Y2,Y3]

names = ['classif', 'blob','moon']

#show datasets
for X, Y in zip(Xs,Ys):
    plt.scatter(X[:, 0], X[:, 1], s=10, c=Y)
    plt.show()
```

### Splitting datasets

```python
val_size = 0.15 #fixed
splitted_data = []
```

```python
def split_val(data, val_size):
    len_val = int( len(data) * val_size )
    len_tt = len(data) - len_val
    np.random.shuffle(data)
    return (data[:len_tt], data[len_tt:])

for X, Y in zip(Xs,Ys):
    data = list(zip(X,Y))
    splitted_data.append( split_val( data, val_size ) )
    
    
for test_train_set, validation_set in splitted_data:
    print(len(test_train_set), len(validation_set))
    
```

### Optimization & Score

```python
dic = {}
accuracy_scores = {}
best_params = {}

for data, name in zip(splitted_data, names):
    test_train_set, validation_set = data
    v_X = []
    v_Y = []
    for x,y in validation_set:
        v_X.append(x)
        v_Y.append(y)
        
    dic = {}
    for ker in ["linear", "poly", 'rbf']:
        for c in np.arange(0.1,10,0.1):
            c = c.round(1)
            clf = svm.SVC(C= c, kernel= ker, gamma='scale')
            ID = ker+'|'+str(c)
            dic[ID] = cross_validation(v_X, v_Y,10, clf)[0]

    
    #print(name, max_param, dic)
        
    max_param = max(dic, key=dic.get)
    max_ker = max_param.split('|')[0]
    max_c = float( max_param.split('|')[1] )
    
    tt_X = []
    tt_Y = []
    for x,y in test_train_set:
        tt_X.append(x)
        tt_Y.append(y)
    
    clf = svm.SVC(C= max_c, kernel= max_ker, gamma='scale')
    cv_score, tests = cross_validation(tt_X, tt_Y,10, clf) 
    
    print('DATASET: '+name+", C_OPT: "+str(max_c)+", KERNEL_OPT: "+max_ker+', SCORE:', cv_score)
    print('\n')
    
    accuracy_scores[name] = tests
    best_params[name] = (max_ker, max_c)
    
```

### Visuals

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.boxplot(accuracy_scores.values())
ax.set_xticklabels(accuracy_scores.keys(), rotation= 'vertical')
plt.title("Optimal accuracy scores")
plt.show()
```

```python
test_perc = 0.2

for X, y, name in zip(Xs,Ys, names):
    X = np.array(X)
    y = np.array(y)
    
    data = list(zip(X,y))
    np.random.shuffle(data)
    
    test_size = int( len(data) * test_perc ) 
    train_size = len(data) - test_size
    
    train = [data[i] for i in range(train_size)]
    test = [data[i] for i in range(test_size)]
    
    X_train, y_train = zip(*train)
    X_test, y_test = zip(*test)
    
    X_train = np.array(X_train)
    X_test = np.array(X_test)    
    y_train = np.array(y_train)
    y_test = np.array(y_test)

    best_ker, best_c = best_params[name]
    print(name)
    
    for kernel in ('linear', 'rbf', 'poly'):

        
        clf = svm.SVC(C= best_c, kernel= kernel, gamma='scale')
        clf.fit(X_train, y_train)
        
        plt.figure(figsize=(12,8))
        plt.clf()
        plt.scatter(X[:, 0], X[:, 1], c=y, zorder=10, cmap=plt.cm.Paired,
                    edgecolor='k', s=20)
        # Circle out the test data
        plt.scatter(X_test[:, 0], X_test[:, 1], s=80, facecolors='none',
                    zorder=10, edgecolor='k')

        plt.axis('tight')
        x_min = X[:, 0].min()
        x_max = X[:, 0].max()
        y_min = X[:, 1].min()
        y_max = X[:, 1].max()

        XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
        Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])

        # Put the result into a color plot
        Z = Z.reshape(XX.shape)
        plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
        plt.contour(XX, YY, Z, colors=['k', 'k', 'k'],
                    linestyles=['--', '-', '--'], levels=[-.5, 0, .5])

        plt.title(kernel)

    plt.show()
```

Support vectors are very important to measure and maximize the distance between the separator and the first point of each classes, also called margin.

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

## Real data

### Import dataset

```python
data = arff.loadarff('data/messidor_features.arff')
X = pd.DataFrame(data[0])

X
```

```python
y = X['Class']
for i in range(len(y)):
    y[i] = y[i].decode('utf8')
    
X.drop(columns= 'Class', inplace = True) 

X = X.values
y = y.values

X,y
```

### split validation/test-train

```python
data = list(zip(X,y))
tt_data, v_data = split_val( data, val_size )

print(len(tt_data), len(v_data))
```

### Optimization & Scoring

```python
v_X, v_y = zip(*v_data)
tt_X, tt_y = zip(*tt_data)

dic = {}
for ker in ["linear", "poly", 'rbf']:
    for c in np.arange(0.1,10,0.1):
        c = c.round(1)
        clf = svm.SVC(C= c, kernel= ker, gamma='scale')
        ID = ker+'|'+str(c)
        dic[ID] = cross_validation(v_X, v_y,10, clf)[0]

max_param = max(dic, key=dic.get)
max_ker = max_param.split('|')[0]
max_c = float( max_param.split('|')[1] )
    
clf = svm.SVC(C= max_c, kernel= max_ker, gamma='scale')
cv_score, tests = cross_validation(tt_X, tt_y,10, clf) 
    
print("C_OPT: "+str(max_c)+", KERNEL_OPT: "+max_ker+', SCORE:', cv_score)
print('\n')
    
accuracy_scores = tests
best_params = (max_ker, max_c)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.boxplot(accuracy_scores)
ax.set_xticklabels(["Diabetic Retinopathy"])
plt.title("Optimal accuracy scores")
plt.show()
```

What are the optimal parameters (kernel and C) for the simulated data?  And for the DiabeticRetinopathy data?  Do you have an intuition, why some kernels are more suitable than others?

First we can notice there's a significant lower (average) accuracy on real data.  
We can see that each kernel have its specialty as each kernel has been selected once for a distinct dataset.

DATASET: blob, C_OPT: 0.2, KERNEL_OPT: linear  
DATASET: moon, C_OPT: 0.1, KERNEL_OPT: rbf  



Kernels work better when they're used on datasets that fit their initial motivation.
