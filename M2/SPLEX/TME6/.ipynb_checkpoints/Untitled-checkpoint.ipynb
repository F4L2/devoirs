```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn import svm

from scipy.io import arff
```

```python
data = arff.loadarff('data/messidor_features.arff')
X = pd.DataFrame(data[0])

y = X['Class']
for i in range(len(y)):
    y[i] = y[i].decode('utf8')
    
X.drop(columns= 'Class', inplace = True) 

X = X.values
y = y.values
```

```python
def split_data(data, k):
    chunks = [None]*k
    c_size = int(len(data) / k)
    for i in range(k):
        chunks[i] = data[i*c_size:(i+1)*c_size]
    return chunks

def cross_validation(X,Y, k, model):
    data = list(zip(X,Y))
    chunks = split_data(data,k)
    scores = []
    for i in range(len(chunks)):
        test_data = chunks[i].copy()
        test_X, test_y = zip(*test_data)

        train_chunks = [chunks[c].copy() for c in range(len(chunks)) if(c != i)]
        train_data = []
        for tc in train_chunks:
            train_data.extend(tc)
            
        train_X, train_y = zip(*train_data)

        model.fit(train_X, train_y)
        prediction = model.predict(test_X)

        score = sum( [1 if(y == y_est) else 0 for y,y_est in zip(test_y, prediction)] ) / len(test_y)
        scores.append(score)
    
    return np.mean(scores) 
```

```python
dic = {}
for ker in ["linear", "poly", 'rbf']:
    dic[ker] = []
    for c in np.arange(0.1,10,0.1):
        c = c.round(1)
        
        print(ker, c)
        clf = svm.SVC(C= c, kernel= ker, gamma='scale')
        
        dic[ker].append( cross_validation(X,y,10, clf) )

dic
```

```python
for k,v in dic.items():
    dic[k] = np.array(dic[k])
    print(k + ", C: "+ str(np.argmax(dic[k]) * 0.1 + 0.1) + " max: "+ str(max(dic[k])))
```

```python
from sklearn.manifold.t_sne import TSNE
from sklearn.neighbors.classification import KNeighborsClassifier

def plot_svm(model, X):
    X_Train_embedded = TSNE(n_components=2).fit_transform(X)
    y_predicted = model.predict(X)

    # create meshgrid
    resolution = 100 # 100x100 background pixels
    X2d_xmin, X2d_xmax = np.min(X_Train_embedded[:,0]), np.max(X_Train_embedded[:,0])
    X2d_ymin, X2d_ymax = np.min(X_Train_embedded[:,1]), np.max(X_Train_embedded[:,1])
    xx, yy = np.meshgrid(np.linspace(X2d_xmin, X2d_xmax, resolution), np.linspace(X2d_ymin, X2d_ymax, resolution))

    # approximate Voronoi tesselation on resolution x resolution grid using 1-NN
    background_model = KNeighborsClassifier(n_neighbors=1).fit(X_Train_embedded, y_predicted) 
    voronoiBackground = background_model.predict(np.c_[xx.ravel(), yy.ravel()])
    voronoiBackground = voronoiBackground.reshape((resolution, resolution))

    #plot
    plt.contourf(xx, yy, voronoiBackground)
    plt.scatter(X_Train_embedded[:,0], X_Train_embedded[:,1], c=y)
    plt.show()
    

plot_svm(clf, X)
```

```python
#generate simulated datasets
X1, Y1 = make_classification(n_samples=10000, n_features=2, n_redundant=0, n_informative=2,n_clusters_per_class=1)
X2, Y2 = make_blobs(n_samples=10000, n_features=2, centers=3)
X3, Y3 = make_moons(n_samples=10000, shuffle=True, noise=None, random_state=None)

Xs = [X1,X2,X3]
Ys = [Y1,Y2,Y3]

#show datasets
for X, Y in zip(Xs,Ys):
    plt.scatter(X[:, 0], X[:, 1], s=10, c=Y)
    plt.show()
```

```python
def split_val(data, val_size):
    len_val = int( len(data) * val_size )
    len_tt = len(data) - len_val
    np.random.shuffle(data)
    return (data[:len_tt], data[len_tt:])



val_size = 0.2
splitted_data = []



for X, Y in zip(Xs,Ys):
    data = list(zip(X,Y))
    splitted_data.append( split_val( data, val_size ) )
    
    
for test_train_set, validation_set in splitted_data:
    print(len(test_train_set), len(validation_set))
    
    
```
